# Week 5: Loops

## 5.1 Repetition in Code

So far, we have learned how to write a variety tasks in code using statements, and how to control the flow of execution through that code based on variables using conditionals. In many problems, we will have to repeat a block of code mutliple times. Say we wanted to compute the average grade of a student based on several assignments. One way to do this would be to repeatedly write out the addition once per grade:

```python
grades = [99,94,87,90,100]

sum = grades[0]
sum = sum + grades[1]
sum = sum + grades[2]
sum = sum + grades[3]
sum = sum + grades[4]

average = sum / 5
print(average)
```

This approach is inefficient for several reasons. It takes much longer to write because you have to add one line for every item in the list. If the number of items in the list changes, you must modify the code to get the correct result. This also introduces opportunity for errors as you may forget to add an element or double-count it. 

## 5.2 `while` Loops

A much more efficient way to do this is to use a loop. A loop is a control structure that repeats code a number of times. It can repeat a fixed number of times or be referenced to a variable using a conditional statement. Python has two types of loops, the simpler being the `while` loop. The `while` loop contains a conditional statement, and repeatedly executes a block of code while that statement is true, then continues down the code afterwards:

```python
x = 0
while(x < 10):
    print(x)
    x = x + 1
print("loop done")
```

Like conditional statements, loops enclose an indented section of code, with unindented code being outside the loop and executed after it finishes. When using while loops, it is possible to make the loop run forever. In this case, the loop will repeat until the interpreter is stopped. This is useful in some cases, but also easy to do accidentally (N.B. If you run the following cell, you will have to restart the interpreter to stop it and run other code). Try to analyze why the loop never exits:

```python
x = 0
while(x < 10):
    print(x)
    x = x - 1
print("loop never done")
```

The following code does the same calculation as our grade example earlier. Notice the use of a variable to keep track of index in the list:

```python
grades = [99,94,87,90,100]
i = 0
sum = 0
while(i < len(grades)):
    sum = sum + grades[i]
    i = i + 1

average = sum / len(grades)
print(average)
```

In this example, we do not have to manually check the length of the list, and the code will still work if another element is added or removed.

## 5.3 `for` Loops

The `for` loop is the second type of loop in Python. Under the hood, it is really a `while` loop, but it covers the common case where we want a loop to keep track of an index variable and so it is a helpful syntactical shortcut.(There are many such shortcuts in programming languages. They are called "syntactic sugar" because while they do not implement new functionality, they make the programmer's life much easier.) The `for` loop is used to iterate over a collection such as a `list`, executing once per item. The most common use of the `for` loop is in conjuction with `range()` to execute code a specific number of times:

```python
n = 10
for i in range(0,n):
    print(i)
```

Recall that `range()` produces a list of integers between the two arguments. The for loop syntax creates a variable, which in this case we name `i`. Each time the loop executes, `i` takes the value of the next thing in the collection; the first time the loop executes `i` is `0`, the next `i` is `1`, and so on. The collection can also be a `list` that we have stored in a variable. The following cell shows the average grade example, rewritten using a `for` loop:

```python
grades = [99,94,87,90,100]
sum = 0
for i in grades:
    sum = sum + i

average = sum / len(grades)
print(average)
```

The `for` loop is typically useful when you want to do an operation on all items in a list, or when you want to do an operation a specific number of times. Recall, however, that `for` loops can be rewritten with a `while` loop.
