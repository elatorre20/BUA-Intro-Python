## 4.1 Conditional Statements

In general, code (like human language) executes in a linear fashion, from top to bottom. In all of the examples we have looked at so far, the interpeter starts at the first line of the cell and runs them all in the order we type them. However, for many problems we might want to program, this is insufficient. We need some way to allow the code to change its execution based on conditions. For example, say that we are writing code to run on a thermostat. The commands sent to the heater and air conditioner will have to change depending on the value of the measured temperature. In programming, this is called a conditional, as it depends on the conidition of some part of the program. In order to handle conditional behavior like this, Python provides us with a variety of conditional expressions.

Recall the `bool` primitive type, which we discussed in week 1. A `bool` can have the value `True` or `False`. Conditional statments are composed of operations that can be evaluated to a `bool`. The following comparison operators are used in Python:

| Operator             |Functionality                               |
|----------------------|-------------------------------------       |
| `==`             | Equivalence (equality, in the case of numbers) |
| `>`              | Greater than                                   |
| `>=`             | Greater or equal to                            |
| `<`              | Less than                                      |
| `<=`             | Less than or equal                             |

The expressions will evaluate to `True` if the comparison is true for the provided values, and `False` if it is not. This allows us to check whether conditions are true:

```python
x = 10
y = 5
print(x > y)
print(x <= y)
```

We can also combine comparison operators or `bool` values using logical operators. The following logical operators are used in Python:

| Operator  | Functionality                              |
|-----------|--------------------------------------------|
| `and`     | `True` if both conditions are true         |
| `or`      | `True` if either conidition is true        |
| `not`     | Inverts the truth value of a condition     |

```python
a = True
b = False
print(a and b)
print(a or b)
print(not b)
```

Like the mathematical operators, comparison and logical operators can be combined in a single statement: 

```python
print(x <= y and x == x or b and not a)
```

## 4.2 `if`/`elif`/`else` Statements

Now that we have a way of generating boolean expressions from values, we need a way to change the execution of code based on these expressions. Control statements allow blocks of code to be executed or not based on the value of a boolean expression. 

An `if` statement executes a block if the condition entered is `True`, and does not if it is `False`. Note the indentation after the `if` statement. In Python, blocks of code are separated by indenation. This means that anything in the indent will only be executed if the `if` statement is true, but anything not indented is not and will execute regardless:

```python
x = 1
y = 2
if(x > y):
    print("x is bigger")
if(x < y):
    print("x is smaller")
print("this will always print")
```

Like most statements, `if` statements can be nested:

```python
z = 2
if(z > x):
    print("x is bigger than y")
    if(z < y):
        print("z is smaller than y")
```

All `if` statements are evaluated independently, so the execution of one does not affect the execution of subsequent ones. In some cases, we may want to execute only one of a set of options. We can do this using `elif` blocks. Once one statement in an `if`/`elif` block has been executed, all subsequent `elif` statements are ignored:

```python
temp = 28

if(temp > 30):
    print("hot")
if(temp > 20):
    print("comfortable")
elif(temp > 10):
    print("cold")
```

After an `if` or at the end of a set of `if`/`elif` blocks, we can provide a default option `else` that will execute if none of the previous blocks execute:

```python
temp = 0

if(temp > 30):
    print("hot")
if(temp > 20):
    print("comfortable")
elif(temp > 10):
    print("cold")
else:
    print("freezing")
```

## 4.3 `match` Statements

Although the `if`/`elif`/`else` statements are sufficient for almost all code control, Python provides an additional type of control statement called a pattern match for convenience. The pattern match behaves similarly to a set of `if`/`elif`/`else` statements, taking the statement to be matched and searching the cases top to bottom until either finding a match and executing that case or executing the default case:

```python
num = 3.14

match(num):
    case 2.72:
        print('e')
    case 3.14:
        print('pi')
    case 9.8:
        print('g')
    case _:
        print("not a fundamental constant")
```
