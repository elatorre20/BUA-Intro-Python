# Week 3: Composite Types

## 3.1 Primitive Types

Last week, we looked at specific examples of data types such as `int`s, `float`s, and `bool`s. At a hardware leve, in RAM, all data are stored as binary numbers and have no type. The same value stored in RAM will have different values in the interpreter depending on what type it is interpreted as. At the level of language syntax, types are a way for the Python interpreter to determine what operations will work on what data and how. For example the `+` operator refers to addition when used with two `int`s or `float`s, concatenation when used with two `string`s, and does not have any functionality when used with one numerical argument and one `string`. 

The examples we discussed last week are all (with the exception of `string`) "primitive types", meaning that they cannot be deconstructed into other types. There are many more things we can do with types, however. By combining primitive types in a specified way, we can create new types that store information in a different way and allow new operations. A good example of this is the `list`. 

## 3.2 Lists

A `list` is an ordered collection of elements of any type. The `list` is composite because it can be decomposed into elements of a different type. `string`s are technically a special kind of `list`, composed of characters. The elements in the list can be of a primitive type like `int`, or they can be themselves of a composite type, for example a `list` of `list`s. All elements in a `list` must be of the same type:

```python
groceries = ['Apples', 'Carrots', 'Chicken', 'Soup']
type(groceries)
```

We can access elements in the list by their index. Index is the place in the list order that an element occupies. Note that indexing in Python (as in most programming languages) begins with 0, so the first element in the list is actually at index 0:

```python
print(groceries[0])
print(groceries[1])
print(groceries[2])
print(groceries[3])
```

An element of the list accessed by index is a statement, and can be used anywhere a statement could be used:

```python
print('Today I will eat ' + groceries[0] + ' and ' + groceries[3])
```

Trying to access an index that is beyond the end of a `list` will produce an error:

```python
print(groceries[4])
```

It is possible to use negative indices to count from the end of a `list`. The index -1 refers to the last item of the `list`, -2 refers to the second-to-last, and so on:

```python
print(groceries[-1])
```

`list`s can be created with elements of any type, including composite types like `list`. Note that in this example, even though the `list`s have different type elements, the `list` of `list`s is allowed because all of its elements are of type `list`:

```python
numbers = [1, 2, 3]
decimals = [3.14, 2.72]
list_of_lists = [groceries, numbers, decimals]
print(list_of_lists[1])
```

In a `list` of `list`s, we can access elements of the inner lists using double indexing:

```python
print(list_of_lists[0][1])
```

This can theoretically be continued for as many levels of `list` as needed. 

We can add items to a `list` with `append()`:

```python
groceries.append("pasta")
print(groceries)
```

`len()` finds the length of the list:

```python
print(len(groceries))
```

There are many other operations that are possible with `list`s. ## 3.3 Dictionaries

The `lis`t is the simplest composite type, but we can build more complex data structures that store more information in different ways. The `list` is just an ordered collection, but often we will want to store more information and relations between elements than just ordering. A dictionary or `dict` stores elements in key-value pairs. Analogous to an actual dictionary, elements of a `dict` are referenced by key rather than by index. This means that elements can be accessed by whatever information stored in the key rather than just an order: 

```python
a_dictionary = {
    'animal': 'any member of the kingdom Animalia, comprising multicellular organisms that have a well-defined shape and usually limited growth, can move voluntarily, actively acquire food and digest it internally, and have sensory and nervous systems that allow them to respond rapidly to stimuli: some classification schemes also include protozoa and certain other single-celled eukaryotes that have motility and animallike nutritional modes.',
    'flower': 'the blossom of a plant.',
    'cat': 'a small domesticated carnivore, Felis domestica or F. catus, bred in a number of varieties.'
}

print(a_dictionary['animal'])
```

Dictionary elements (both key and value) can take any type:

```python
class_times = {
    9 : "history",
    10 : "algebra",
    11 : "lunch"
}

print(class_times[10])
```

Unlike `list`s, it is possible to mix keys or values of different types in the same `dict`. Similar to the `list`, accessing a key that does not exist will produce an error. However, unlike the list we cannot just check the length to determine whether a key exists in a given `dict`. We can use the `in` operator to check for a key in a dictionary. `in` will be `True` if the key is in the dictionary, and `False` if it is not. 

```python
'funny' in a_dictionary
```

`keys()` gets a list of all keys in the dictionary:

```python
a_dictionary.keys()
```

### 3.4 More Complex Dictionaries

We can continue even further with dictionaries by using composite types as values. Nested `dict`s can be used to create relatively complex data structures. Consider the following `dict`, which organizes many data of different types into one structure: 

```python
my_dictionary = {
    'name': 'Boston',
    'population': 654776,
    'area': 89.63,
    'universities': {
        'Boston University': {
            'students':  37557,
            'majors': ['Computer Science', 'Biology', 'History', 'Math']
        },
        'MIT': {
            'students':  11858,
            'majors': ['Computer Science', 'Biology', 'History', 'Math', 'Robotics']
        },
        'Harvard': {
            'students':   25266,
            'majors': ['Law', 'Biology', 'History', 'Math', 'Robotics']
        },    
    }
}
```
